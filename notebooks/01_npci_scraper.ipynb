```python
# Cell 1 — Import Libraries
# These are all the tools our scraper needs

import requests                        # sends HTTP requests to websites
from bs4 import BeautifulSoup          # reads and parses HTML
import pandas as pd                    # handles data in table format
import time                            # adds delays between requests (polite scraping)
import os                              # handles file paths and folders

print("All libraries imported successfully ✓")
```

```python
# Cell 2 — Set Up Project Paths
# This tells Python where to save all scraped data
# os.path.join builds the correct path for Windows automatically

# Get the current notebook folder location
notebook_dir = os.path.dirname(os.path.abspath("01_npci_scraper.ipynb"))

# Build path to data/raw folder
raw_data_path = os.path.join(notebook_dir, "..", "data", "raw")

# Create the folder if it doesn't exist
os.makedirs(raw_data_path, exist_ok=True)

print(f"Notebook folder : {notebook_dir}")
print(f"Raw data folder : {os.path.abspath(raw_data_path)}")
print("Folders ready ✓")
```

```python
# Cell 3 — Test if NPCI website is reachable
# Before scraping we check the site responds to our request
# headers makes our request look like a real browser
# without this, some websites block automated requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

url = "https://www.npci.org.in/what-we-do/upi/upi-ecosystem-statistics"

response = requests.get(url, headers=headers, timeout=15)

print(f"Status code : {response.status_code}")
print(f"Response size : {len(response.content)} bytes")

if response.status_code == 200:
    print("NPCI website is reachable ✓")
else:
    print("Something went wrong ✗")
```
